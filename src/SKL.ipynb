```python
import os
titles=[]
authors=[]
for path, subdirs, files in os.walk("../resources"):
    for name in files:
        filename=os.path.join(path, name)
        titles.append(filename)
        authors.append(filename[filename.index("resources\\")+10:filename.rfind("\\")])
print(", ".join(titles))
print(", ".join(authors))
```

```python
books=[open(title, encoding="utf-8", errors="ignore").read() for title in titles]
titles=[text[:text.index("~~~")] for text in books]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pickle
```

```python

#max_df eliminates words that appear in >50% of the texts
#stop_words eliminates common words from inclusion in the final matrix
vectorizer = TfidfVectorizer(max_df=.5, 
                             stop_words='english')

tfidf_matrix = vectorizer.fit_transform(books)
pickle.dump(titles, open("titles.pickle", "wb"))
pickle.dump(vectorizer, open("vectorizer.pickle", "wb"))
pickle.dump(tfidf_matrix, open("tfidf_matrix.pickle", "wb"))
```

```python
tfidf_matrix=pickle.load(open("tfidf_matrix.pickle", "rb"))
vectorizer=pickle.load(open("vectorizer.pickle", "rb"))
print(titles[-1])
print(tfidf_matrix[-1])
```

```python
#see most distinctive terms in each book
terms=vectorizer.get_feature_names()
top_indices=(-tfidf_matrix[:100]).toarray().argsort()[:, :5]

for index, title in enumerate(titles[:100]):
    top_terms=[terms[t] for t in top_indices[index]]
    print(f"{title}:\n{', '.join(top_terms)}\n\n")
```

```python
#get the cosine similarities per text
from sklearn.metrics.pairwise import linear_kernel
import collections
from nltk.tokenize import word_tokenize
with open("../userChosenText.txt", "r", encoding="utf-8", errors="ignore") as user_text:
#     user_text=user_text.read().lower()
#     user_text=word_tokenize(user_text)
#     user_bow=collections.Counter(user_text).items()
    v=vectorizer.vocabulary_
    new_vectorizer=TfidfVectorizer(stop_words="english", vocabulary=v)
    user_tfidf=new_vectorizer.fit_transform(user_text)
print(user_tfidf)
cosine_similarities = linear_kernel(tfidf_matrix[-1:], tfidf_matrix).flatten()
related_text_indices = cosine_similarities.argsort()
```

```python
related_text_titles = [titles[int(i)] for i in related_text_indices]
related_text_titles=related_text_titles[0:-1]
print(related_text_titles)
```

```python
import csv
with open('../book_tfidf_closeness.csv', 'w+', encoding="utf-8", errors="ignore") as csv_file:
    writer = csv.writer(csv_file)
    for i, title in enumerate(related_text_titles):
        writer.writerow([title, len(related_text_titles)-i])
```
